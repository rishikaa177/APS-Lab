# Minimum Cost Spanning Tree
Problem Description:
A spanning tree of a graph can be defined as a graph with minimal set of edges that connect all vertices. A minimum spanning tree of a graph is a spanning tree of the graph with least weight (where the weight is computed by adding the weights of all the edges in the spanning tree). In general, a graph can have multiple minimum spanning trees. The problem is to find a minimum spanning tree of a graph.

This implementation uses a dictionary to represent the weighted graph, where the vertices are keys, and the values are dictionaries containing their adjacent vertices and their weights.

Please note that this implementation assumes that the input graph is connected and weighted.

First, let's create a simple graph to demonstrate Prim's algorithm:

```python
graph = {
    "A": {"B": 10, "C": 15, "D": 20},
    "B": {"A": 10, "C": 12},
    "C": {"A": 15, "B": 12, "D": 18},
    "D": {"A": 20, "C": 18},
}
```

Now, let's implement Prim's algorithm:

```python
def prim_algorithm(graph):
    min_tree = {}
    visited = set()
    min_tree[list(graph.keys())[0]] = []
    min_weight = 0

    while len(visited) < len(graph):
        min_edge = float('inf')
        next_vertex = None
        for vertex in min_tree:
            for adj_vertex, weight in graph[vertex].items():
                if adj_vertex not in visited:
                    if weight < min_edge:
                        min_edge = weight
                        next_vertex = adj_vertex

        min_tree[next_vertex] = []
        visited.add(next_vertex)
        min_weight += min_edge

    return min_tree, min_weight
```

Now, let's run the Prim's algorithm on the example graph:
Keep in mind that this example uses a simple representation of a graph, but for more complex cases, you might want to use a library such as NetworkX to represent and manipulate graphs.

```python
mst, min_weight = prim_algorithm(graph)
print("Minimum Spanning Tree: ", mst)
print("Minimum Weight: ", min_weight)
```

```python
#run prims algorithm for sample input
def prim_algorith(graph):
  mst_nodes = {}
  visited = set()
  min_weight = 0

  # Start with an arbitrary vertex (e.g., the first one in the graph)
  if not graph:
    return {}, 0
  start_vertex = list(graph.keys())[0]
  visited.add(start_vertex)
  mst_nodes[start_vertex] = [] # Initialize MST with the start vertex

  while len(visited) < len(graph):
    min_edge_weight = float('inf')
    next_node_to_add = None

    # Iterate over all vertices currently in the MST
    for u in visited:
      # Iterate over all neighbors of u
      for v, weight in graph[u].items():
        # If v is not yet in the MST and this edge is lighter than current min_edge_weight
        if v not in visited and weight < min_edge_weight:
          min_edge_weight = weight
          next_node_to_add = v

    # If no edge was found, it means the graph is disconnected
    if next_node_to_add is None:
      break

    # Add the selected node to the MST
    visited.add(next_node_to_add)
    mst_nodes[next_node_to_add] = [] # Add the new node to the MST nodes dict
    min_weight += min_edge_weight

  return mst_nodes, min_weight

# These lines are now outside the function definition, as they should be
mst, min_weight = prim_algorith(graph)
print("Minimum Spanning Tree: ", mst)
print("Minimum Weight: ", min_weight)
```

```python
# implement Kruskal's algorithm
graph = {
    "A": {"B": 6,"D": 1},
    "B": {"A": 6, "C": 5,"D":2,"E":2},
    "C": {"A": 1, "B": 2, "E": 1},
    "D": {"A": 1, "B": 2,"E":1},
    "E": {"B": 2, "C": 1,"D":1}
}

def find(parent,vertex):
  if parent[vertex] != vertex:
    parent[vertex] = find(parent,parent[vertex])
  return parent[vertex]

def union(parent,rank,vertex1,vertex2):
  root1 = find(parent,vertex1)
  root2 = find(parent,vertex2)
  if rank[root1] < rank[root2]:
    parent[root1] = root2
  else:
    parent[root2] = root1
    if rank[root1] == rank[root2]:
      rank[root1] +=1

def kruskal_algorithm(graph):
  edges = []
  for u in graph:
    for v, weight in graph[u].items():
      # To avoid duplicate edges for undirected graphs (e.g., (A,B,6) and (B,A,6))
      # We'll add each edge once. A simple way is to ensure u < v, or just add all and let DSU handle cycles.
      # For this specific graph representation, collecting (u,v,w) and (v,u,w) is fine;
      # sorting and DSU will ensure correctness.
      edges.append((weight, u, v)) # Store as (weight, u, v) for easy sorting by weight

  edges.sort() # Sort edges by weight

  parent = {}
  rank = {}
  mst = []
  min_weight = 0

  for vertex in graph:
    parent[vertex] = vertex
    rank[vertex] = 0

  for w, u, v in edges:
    if find(parent,u) != find(parent,v):
      union(parent,rank,u,v)
      mst.append((u,v,w)) # Store edges in MST as (u,v,weight)
      min_weight += w

  return mst, min_weight

# Call the Kruskal's algorithm and print results outside the function definition
mst, min_weight = kruskal_algorithm(graph)
print("Minimum Spanning Tree: ", mst)
print("Minimum Weight: ", min_weight)
```
